# Avocado Project

# Project Description

This data was downloaded from the Hass Avocado Board website in May of 2018 & compiled into a single CSV. 

The table below represents weekly 2018 retail scan data for National retail volume (units) and price.
Retail scan data comes directly from retailers’ cash registers based on actual retail sales of Hass avocados. 
Starting in 2013, the table below reflects an expanded, multi-outlet retail data set. Multi-outlet reporting includes an aggregation of the following channels: grocery, mass, club, drug, dollar and military. The Average Price (of avocados) in the table reflects a per unit (per avocado) cost, even when multiple units (avocados) are sold in bags. 
The Product Lookup codes (PLU’s) in the table are only for Hass avocados. 
Other varieties of avocados (e.g. greenskins) are not included in this table.

Some relevant columns in the dataset:

•	Date - The date of the observation

•	AveragePrice - the average price of a single avocado

•	type - conventional or organic

•	year - the year

•	Region - the city or region of the observation

•	Total Volume - Total number of avocados sold

•	4046 - Total number of avocados with PLU 4046 sold

•	4225 - Total number of avocados with PLU 4225 sold

•	4770 - Total number of avocados with PLU 4770 sold

```python
def metric_score(clf, x_train,x_test,y_train,y_test, train = True):
    if train:
        y_pred = clf.predict(x_train)
        
        print("\n===============Train Result =============================")
        
        print(f"Accuracy Score: {accuracy_score(y_train, y_pred) * 100:.2f}%")
        
    elif train==False :
        pred = clf.predict(x_test)
        
        print("\n===============Test Result =============================")
        print(f"Accuracy Score: {accuracy_score(y_test, pred) * 100:.2f}%")
                        
        print('\n \n Test Classification Report \n', classification_report(y_test, pred, digits = 2)) ## model confidance/accuracy
        
```

```python
# Checking accuracy for Extra Tree classifier 
ET = ExtraTreesClassifier()
ET.fit(x_train, y_train)

```

```python
# Call the function and pass dataset to check train and test score 
metric_score(ET,x_train,x_test,y_train,y_test,train=True)  #This is for training score 

metric_score(ET,x_train,x_test,y_train,y_test,train=False) # This is for testing score
```

# # Separating features and label Separating features and label for Average Price

```python
x1 = df1.drop("AveragePrice",axis = 1)
y1= df1["AveragePrice"]
```

# Creating the training and testing data sets for average price

```python
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1,y1,test_size=0.2, random_state=42)

print('Size of X_train : ', x1_train.shape)
print('Size of y_train : ', y1_train.shape)
print('Size of X_test : ', x1_test.shape)
print('Size of Y_test : ', y1_test.shape)
```

# Support Vector Machine for price

```python
from sklearn.svm import SVR
svr=SVR()
svr.fit(x1_train, y1_train)
```

```python
from sklearn.model_selection import cross_val_score, RandomizedSearchCV, GridSearchCV

parameters =  { 'kernel' : ['rbf', 'sigmoid'],
                 'gamma' : [0.001, 0.01, 0.1, 1, 'scale'],
                 'tol' : [0.0001],
                 'C': [0.001, 0.01, 0.1, 1, 10, 25] }
svr_grid = GridSearchCV(estimator=svr, param_grid=parameters, cv=10, verbose=4, n_jobs=-1)
svr_grid.fit(x1_train, y1_train)
```

```python
svr = SVR(C=10, gamma=0.1, tol=0.001)
svr.fit(x1_train, y1_train)
print(svr_grid.best_estimator_)
print(svr_grid.best_score_)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, accuracy_score, confusion_matrix
```

```python
cv_svr = svr_grid.best_score_

y_pred_svr_train = svr.predict(x1_train)
r2_score_svr_train = r2_score(y1_train, y_pred_svr_train)

y_pred_svr_test = svr.predict(x1_test)
r2_score_svr_test = r2_score(y1_test, y_pred_svr_test)

rmse_svr = (np.sqrt(mean_squared_error(y1_test, y_pred_svr_test)))

print('CV : {0:.3f}'.format(cv_svr.mean()))
print('R2_score (train) : {0:.3f}'.format(r2_score_svr_train))
print('R2 score (test) : {0:.3f}'.format(r2_score_svr_test))
print('RMSE : {0:.3f}'.format(rmse_svr))
```
